# Principal Component Analysis for Wine Data

```python
#Importing Essential Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv(r'C:\Users\aravi\OneDrive\Desktop\ExcelR Assignments\08. PCA\wine.csv')
```

```python
df=data1.drop(['Type'],axis=1)
df.head()
```

# Exploratory Data Analysis

```python
df.info()
```

```python
df.describe()
```

```python
# Converting into numpy array
data = df.values
data
```

```python
data.shape
```

## Normalizing The Numerical Data

```python
data_norm=scale(data)
```

```python
data_norm
```

```python
pca=PCA()
```

```python
pca_values=pca.fit_transform(data_norm)
```

```python
pca_values
```

```python
pca=PCA(n_components=3)
pca_values=pca.fit_transform(data_norm)
```

```python
pca_values
```

```python
pca.components_
```

```python
# The amount of Variance that is explained by Each PCA 
variance=pca.explained_variance_ratio_
variance
```

```python
# Cumulative variance 
var = np.cumsum(np.round(variance,decimals = 4)*100)
var
```

```python
# Variance plot for PCA components obtained 
plt.plot(var,color="red")
```
